```python
import pandas as pd
import adlfs
from dotenv import load_dotenv
import pandas as pd
import adlfs
from azure.identity import DefaultAzureCredential
import os

def abfss_to_adlfs_path(abfss_path: str) -> str:
    """
    Convert:
      
    """
    if not abfss_path.startswith("abfss://"):
        raise ValueError(f"Expected abfss:// path, got: {abfss_path}")

    without_scheme = abfss_path.replace("abfss://", "", 1)
    container, rest = without_scheme.split("@", 1)

    marker = ".dfs.core.windows.net/"
    if marker not in rest:
        raise ValueError(f"Unexpected ADLS Gen2 URL (missing {marker}): {abfss_path}")

    path = rest.split(marker, 1)[1]
    return f"{container}/{path}".rstrip("/")

def read_csv_folder_from_adls(folder_abfss: str, storage_options: dict, **read_csv_kwargs) -> pd.DataFrame:
    fs = adlfs.AzureBlobFileSystem(**storage_options)

    folder_fs = abfss_to_adlfs_path(folder_abfss).rstrip("/")
    files = fs.glob(folder_fs + "/*.csv")

    if not files:
        raise FileNotFoundError(f"No CSV files found under: {folder_abfss}")

    dfs = []
    for f in files:
        with fs.open(f, "rb") as fp:
            dfs.append(pd.read_csv(fp, **read_csv_kwargs))

    return pd.concat(dfs, ignore_index=True)

def main():
    load_dotenv()

    ACCOUNT_NAME = "chevrondatalake"
    CONTAINER_NAME = "produced"
    DIRECTORY_PATH = "HES/MOC/MOC_GUG"

    base_path = f"abfss://{CONTAINER_NAME}@{ACCOUNT_NAME}.dfs.core.windows.net/{DIRECTORY_PATH}/"

    storage_options = {
        "account_name": ACCOUNT_NAME,
        "anon": False,   # uses your Azure login / environment auth
    }

    detail_path = base_path + "DETAIL/"
    hierarchy_path = base_path + "HIERARCHY/"

    pdf_detail = read_csv_folder_from_adls(detail_path, storage_options, dtype=str)
    pdf_hierarchy = read_csv_folder_from_adls(hierarchy_path, storage_options, dtype=str)

    print("DETAIL:", pdf_detail.shape)
    print("HIERARCHY:", pdf_hierarchy.shape)

    total_moc = pdf_detail.merge(
        pdf_hierarchy,
        how="left",
        on="DEPARTMENTID",
        suffixes=("_d", "_h")
    )

    return total_moc

# In notebook:
total_moc = main()
```

### Merge columns "MOC_CANCELED_DT" and "MOC_CANCELED_MOC_DT

```python
import pandas as pd

def add_moc_canceled_date(df_MOC_gor: pd.DataFrame) -> pd.DataFrame:
    df_MOC_gor = df_MOC_gor.copy()

    required_cols = ["MOC_CANCELED_DT", "MOC_CANCELED_MOC_DT"]
    for c in required_cols:
        if c not in df_MOC_gor.columns:
            df_MOC_gor[c] = pd.NA

    # ✅ Force BOTH to be timezone-aware UTC
    d1 = pd.to_datetime(df_MOC_gor["MOC_CANCELED_DT"], errors="coerce", utc=True)
    d2 = pd.to_datetime(df_MOC_gor["MOC_CANCELED_MOC_DT"], errors="coerce", utc=True)

    # Pick latest
    latest = d1.copy()
    mask_take_d2 = d2.notna() & (d1.isna() | (d2 > d1))
    latest[mask_take_d2] = d2[mask_take_d2]

    # Output string in UTC ISO format (Z)
    df_MOC_gor["MOC_CANCELED_DT"] = latest.dt.strftime("%Y-%m-%dT%H:%M:%SZ").fillna("")
    df_MOC_gor = df_MOC_gor.drop(columns=["MOC_CANCELED_MOC_DT"])

    return df_MOC_gor
df_MOC_gor = add_moc_canceled_date(df_MOC_gor)
```

### Upload to DB

```python
import pandas as pd
import numpy as np
from neo4j import GraphDatabase
import os



URI = os.getenv("NEO4J_URI")
USERNAME = os.getenv("NEO4J_USERNAME")
PASSWORD = os.getenv("NEO4J_PASSWORD")
DATABASE = os.getenv("NEO4J_DATABASE")

BATCH_SIZE = 9000

DATETIME_COLS = [
    'draftDate', 'dueDate','pendingFirstApprovalDate', 'approvedToProceedDate',
    'pendingSecondApprovalDate','approvedToMakeChangeDate', 'pendingStartupApprovalDate',
    'approvedForStartupDate', "mocInOperationDate", 'pendingClosureDate', "mocClosedDate",
    'onHoldDate','cancelledProposedCrDate',  "mocCanceledDate", "desiredStartupDate","approvedToProceedDate",
"cancelledProposedCrDate","desiredStartupDate","desiredStartupDate","draftDate","dueDate","emergencyDate","lastStatusChangeDate","mocCanceledDate",
"mocClosedDate","mocInOperationDate","onHoldDate","pendingClosureDate","pendingFirstApprovalDate","pendingSecondApprovalDate",
"pendingStartupApprovalDate"
]

INT_COLS = ['daysInDraft', 'daysInPendingFirstApproval',  "daysInApprovedToProceed",
     'daysInPendingSecondApproval','daysInApprovedToMakeChange', "daysInPendingStartup",
    "daysInApprovedForStartup", "daysInOperation", "daysPendingClosure",
    'daysUntilDue', 'isOverdue', 'holdCount', "mocPlacedOnHoldCount",'daysOverdue', "daysStagnant",'authorizedDurationDays',
            'parentMocId', 'programNumber',"sbuIdentifier",'afeCost',
]

cypher = """
UNWIND $rows AS row
CREATE (n:Moc)
SET n += row
SET n.sourceFile = ["moc_impact_table"]
SET n.siteId = "5284"
SET n.siteName = "Gorgon"
"""

driver = GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD))

def upload_batch(session, batch):
    session.run(cypher, rows=batch).consume()

def chunked(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i+n]

# -----------------------------
# INPUT: your DataFrame
# -----------------------------
df = df_MOC_gor.copy()
df.columns = df.columns.astype(str).str.strip()

# 2) Trim strings ONLY in object (string-ish) columns to avoid dtype pollution
obj_cols = df.select_dtypes(include=["object"]).columns
df[obj_cols] = df[obj_cols].apply(lambda s: s.map(lambda v: v.strip() if isinstance(v, str) else v))

# 3) Normalize null tokens ONLY in object columns
NULL_TOKENS = {"", "-", "--", "---"}
df[obj_cols] = df[obj_cols].apply(
    lambda s: s.map(lambda v: None if (isinstance(v, str) and v.strip() in NULL_TOKENS) else v)
)


#for col in DATETIME_COLS:
 #   if col in df.columns:
 #       parsed = pd.to_datetime(df[col], errors="coerce", utc=True)
 #       df[col] = parsed.dt.strftime("%Y-%m-%dT%H:%M:%SZ").where(parsed.notna(), None)
       # df[col] = parsed.astype(object).where(parsed.notna(), None)
 #       df[col] = df[col].where(parsed.notna(), None)  # NaT -> None

for col in DATETIME_COLS:
    if col in df.columns:
        parsed = pd.to_datetime(df[col], errors="coerce", utc=True)
        df[col] = parsed.dt.to_pydatetime()  # Python datetime objects
        df[col] = pd.Series(df[col]).where(parsed.notna(), None)

        

# 5) Add "MOC" prefix to mocId (if present)
if "mocId" in df.columns:
    def add_moc_prefix(x):
        if x is None:
            return None
        s = str(x).strip()
        if not s:
            return None
        return s if s.startswith("MOC") else f"MOC{s}"
    df["mocId"] = df["mocId"].map(add_moc_prefix)

# 6) Convert any remaining NaN/NaT to None (safe for Neo4j)
df = df.replace({np.nan: None})
df = df.where(pd.notna(df), None)



for col in INT_COLS:
    if col in df.columns:
        # 1) Convert anything numeric-looking to numbers; bad values like 'N' -> NaN
        s = pd.to_numeric(df[col], errors="coerce")

        # 2) OPTIONAL: if you only want whole numbers, drop decimals (12.5 -> NaN)
        # remove this line if decimals are valid for your column
        s = s.where(s.isna() | (s % 1 == 0), np.nan)

        # 3) Convert to nullable integer (supports missing values as <NA>)
        df[col] = s.astype("Int64")


rows = df.to_dict(orient="records")

# 10) Upload in batches
with driver.session(database=DATABASE) as session:
    for batch in chunked(rows, BATCH_SIZE):
        upload_batch(session, batch)

driver.close()
print("Upload complete ✅ (integers enforced in payload).")

```

```python
import pandas as pd

def add_moc_canceled_date(df_MOC_WSU : pd.DataFrame) -> pd.DataFrame:
    df_MOC_WSU = df_MOC_WSU.copy()

    required_cols = ["MOC_CANCELED_DT", "MOC_CANCELED_MOC_DT"]
    for c in required_cols:
        if c not in df_MOC_WSU.columns:
            df_MOC_WSU[c] = pd.NA

    # ✅ Force BOTH to be timezone-aware UTC
    d1 = pd.to_datetime(df_MOC_WSU ["MOC_CANCELED_DT"], errors="coerce", utc=True)
    d2 = pd.to_datetime(df_MOC_WSU ["MOC_CANCELED_MOC_DT"], errors="coerce", utc=True)

    # Pick latest
    latest = d1.copy()
    mask_take_d2 = d2.notna() & (d1.isna() | (d2 > d1))
    latest[mask_take_d2] = d2[mask_take_d2]

    # Output string in UTC ISO format (Z)
    df_MOC_WSU["MOC_CANCELED_DT"] = latest.dt.strftime("%Y-%m-%dT%H:%M:%SZ").fillna("")
    df_MOC_WSU = df_MOC_WSU.drop(columns=["MOC_CANCELED_MOC_DT"])

    return df_MOC_WSU
df_MOC_WSU = add_moc_canceled_date(df_MOC_WSU)
```
