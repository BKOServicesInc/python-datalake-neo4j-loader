### Merge columns "MOC_CANCELED_DT" and "MOC_CANCELED_MOC_DT

```python
import pandas as pd

def add_moc_canceled_date(df_MOC_gor: pd.DataFrame) -> pd.DataFrame:
    df_MOC_gor = df_MOC_gor.copy()

    required_cols = ["MOC_CANCELED_DT", "MOC_CANCELED_MOC_DT"]
    for c in required_cols:
        if c not in df_MOC_gor.columns:
            df_MOC_gor[c] = pd.NA

    # ✅ Force BOTH to be timezone-aware UTC
    d1 = pd.to_datetime(df_MOC_gor["MOC_CANCELED_DT"], errors="coerce", utc=True)
    d2 = pd.to_datetime(df_MOC_gor["MOC_CANCELED_MOC_DT"], errors="coerce", utc=True)

    # Pick latest
    latest = d1.copy()
    mask_take_d2 = d2.notna() & (d1.isna() | (d2 > d1))
    latest[mask_take_d2] = d2[mask_take_d2]

    # Output string in UTC ISO format (Z)
    df_MOC_gor["MOC_CANCELED_DT"] = latest.dt.strftime("%Y-%m-%dT%H:%M:%SZ").fillna("")
    df_MOC_gor = df_MOC_gor.drop(columns=["MOC_CANCELED_MOC_DT"])

    return df_MOC_gor
df_MOC_gor = add_moc_canceled_date(df_MOC_gor)
```

### Convert to Neo4j naming convention

```python
import re

def to_camel_case(col: str) -> str:
    col = str(col).strip()
    col = re.sub(r"[\s\-]+", "_", col)
    parts = [p for p in col.split("_") if p]
    if not parts:
        return col
    first = parts[0].lower()
    rest = "".join(p[:1].upper() + p[1:].lower() for p in parts[1:])
    return first + rest

# 1) camelCase rename
df_MOC_gor = df_MOC_gor.rename(columns=lambda c: to_camel_case(c))

# 2) then change Dt suffix -> Date (only at the end of the name)
df_MOC_gor = df_MOC_gor.rename(columns=lambda c: re.sub(r"Dt$", "Date", c))
df_MOC_gor = df_MOC_gor.rename(columns={"daysStagnanet": "daysStagnant"})
```

### Upload to DB

```python
import pandas as pd
import numpy as np
from neo4j import GraphDatabase
import os

URI = os.getenv("NEO4J_URI")
USERNAME = os.getenv("NEO4J_USERNAME")
PASSWORD = os.getenv("NEO4J_PASSWORD")
DATABASE = os.getenv("NEO4J_DATABASE")

BATCH_SIZE = 9000

DATETIME_COLS = [
    "crDraftDate", "mocDueDate", "mocPendingFirstApprovalDate", "mocApprovedToProceedDate",
    "mocPendingSecondApprovalDate", "mocApprovedToMakeChangeDate", "mocPendStartupApprovalDate",
    "mocApprovedForStartupDate", "mocInOperationDate", "mocPendingClosureDate", "mocClosedDate",
    "mocOnHoldDate", "mocCanceledProposedCrDate", "mocCanceledDate", "desiredStartupDate",
]

INT_COLS = [
    "daysInCrDraft", "daysInPendFirstApproval", "daysInApprovedToProceed",
    "daysInPendSecondApprover", "daysInApprToMakeChange", "daysInPendingStartup",
    "daysInApprovedForStartup", "daysInOperation", "daysPendingClose",
    "mocDaysTillDue", "mocOverdue", "mocPlacedOnHoldCount", "mocDaysOverdue","daysStagnant","authDurationTempChange",
    "mocParentId","mocProgramNum","sbuIdentifier","mocAfeCost"
]

cypher = """
UNWIND $rows AS row
CREATE (n:Moc)
SET n += row
SET n.sourceFile = ["moc_impact_table"]
SET n.siteId = "5284"
SET n.siteName = "Gorgon"
"""

driver = GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD))

def upload_batch(session, batch):
    session.run(cypher, rows=batch).consume()

def chunked(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i+n]

# -----------------------------
# INPUT: your DataFrame
# -----------------------------
df = df_MOC_gor.copy()
df.columns = df.columns.astype(str).str.strip()

# 2) Trim strings ONLY in object (string-ish) columns to avoid dtype pollution
obj_cols = df.select_dtypes(include=["object"]).columns
df[obj_cols] = df[obj_cols].apply(lambda s: s.map(lambda v: v.strip() if isinstance(v, str) else v))

# 3) Normalize null tokens ONLY in object columns
NULL_TOKENS = {"", "-", "--", "---"}
df[obj_cols] = df[obj_cols].apply(
    lambda s: s.map(lambda v: None if (isinstance(v, str) and v.strip() in NULL_TOKENS) else v)
)


for col in DATETIME_COLS:
    if col in df.columns:
        parsed = pd.to_datetime(df[col], errors="coerce", utc=True)
        df[col] = parsed.dt.strftime("%Y-%m-%dT%H:%M:%SZ").where(parsed.notna(), None)
        df[col] = parsed.astype(object).where(parsed.notna(), None)
        

# 5) Add "MOC" prefix to mocId (if present)
if "mocId" in df.columns:
    def add_moc_prefix(x):
        if x is None:
            return None
        s = str(x).strip()
        if not s:
            return None
        return s if s.startswith("MOC") else f"MOC{s}"
    df["mocId"] = df["mocId"].map(add_moc_prefix)

# 6) Convert any remaining NaN/NaT to None (safe for Neo4j)
df = df.replace({np.nan: None})
df = df.where(pd.notna(df), None)



for col in INT_COLS:
    if col in df.columns:
        # 1) Convert anything numeric-looking to numbers; bad values like 'N' -> NaN
        s = pd.to_numeric(df[col], errors="coerce")

        # 2) OPTIONAL: if you only want whole numbers, drop decimals (12.5 -> NaN)
        # remove this line if decimals are valid for your column
        s = s.where(s.isna() | (s % 1 == 0), np.nan)

        # 3) Convert to nullable integer (supports missing values as <NA>)
        df[col] = s.astype("Int64")


rows = df.to_dict(orient="records")

# 10) Upload in batches
with driver.session(database=DATABASE) as session:
    for batch in chunked(rows, BATCH_SIZE):
        upload_batch(session, batch)

driver.close()
print("Upload complete ✅ (integers enforced in payload).")

```

```python
import pandas as pd

def add_moc_canceled_date(df_MOC_WSU : pd.DataFrame) -> pd.DataFrame:
    df_MOC_WSU = df_MOC_WSU.copy()

    required_cols = ["MOC_CANCELED_DT", "MOC_CANCELED_MOC_DT"]
    for c in required_cols:
        if c not in df_MOC_WSU.columns:
            df_MOC_WSU[c] = pd.NA

    # ✅ Force BOTH to be timezone-aware UTC
    d1 = pd.to_datetime(df_MOC_WSU ["MOC_CANCELED_DT"], errors="coerce", utc=True)
    d2 = pd.to_datetime(df_MOC_WSU ["MOC_CANCELED_MOC_DT"], errors="coerce", utc=True)

    # Pick latest
    latest = d1.copy()
    mask_take_d2 = d2.notna() & (d1.isna() | (d2 > d1))
    latest[mask_take_d2] = d2[mask_take_d2]

    # Output string in UTC ISO format (Z)
    df_MOC_WSU["MOC_CANCELED_DT"] = latest.dt.strftime("%Y-%m-%dT%H:%M:%SZ").fillna("")
    df_MOC_WSU = df_MOC_WSU.drop(columns=["MOC_CANCELED_MOC_DT"])

    return df_MOC_WSU
df_MOC_WSU = add_moc_canceled_date(df_MOC_WSU)
```

```python
import re

def to_camel_case(col: str) -> str:
    col = str(col).strip()
    col = re.sub(r"[\s\-]+", "_", col)
    parts = [p for p in col.split("_") if p]
    if not parts:
        return col
    first = parts[0].lower()
    rest = "".join(p[:1].upper() + p[1:].lower() for p in parts[1:])
    return first + rest

# 1) camelCase rename
df_MOC_WSU  = df_MOC_WSU .rename(columns=lambda c: to_camel_case(c))

# 2) then change Dt suffix -> Date (only at the end of the name)
df_MOC_WSU  = df_MOC_WSU .rename(columns=lambda c: re.sub(r"Dt$", "Date", c))
df_MOC_WSU = df_MOC_WSU.rename(columns={"daysStagnanet": "daysStagnant"})
```

```python
import pandas as pd
import numpy as np
from neo4j import GraphDatabase




URI = os.getenv("NEO4J_URI")
USERNAME = os.getenv("NEO4J_USERNAME")
PASSWORD = os.getenv("NEO4J_PASSWORD")
DATABASE = os.getenv("NEO4J_DATABASE")
BATCH_SIZE = 9000

DATETIME_COLS = [
    "crDraftDate", "mocDueDate", "mocPendingFirstApprovalDate", "mocApprovedToProceedDate",
    "mocPendingSecondApprovalDate", "mocApprovedToMakeChangeDate", "mocPendStartupApprovalDate",
    "mocApprovedForStartupDate", "mocInOperationDate", "mocPendingClosureDate", "mocClosedDate",
    "mocOnHoldDate", "mocCanceledProposedCrDate", "mocCanceledDate", "desiredStartupDate",
]

INT_COLS = [
    "daysInCrDraft", "daysInPendFirstApproval", "daysInApprovedToProceed",
    "daysInPendSecondApprover", "daysInApprToMakeChange", "daysInPendingStartup",
    "daysInApprovedForStartup", "daysInOperation", "daysPendingClose",
    "mocDaysTillDue", "mocOverdue", "mocPlacedOnHoldCount", "mocDaysOverdue","daysStagnant","authDurationTempChange",
    "mocParentId","mocProgramNum","sbuIdentifier","mocAfeCost"
]

cypher = """
UNWIND $rows AS row
CREATE (n:Moc)
SET n += row
SET n.sourceFile = ["moc_impact_table"]
SET n.siteId = "3388"
SET n.siteName = "WSU"
"""



driver = GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD))

def upload_batch(session, batch):
    session.run(cypher, rows=batch).consume()

def chunked(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i+n]

# -----------------------------
# INPUT: your DataFrame
# -----------------------------
df = df_MOC_WSU.copy()
df.columns = df.columns.astype(str).str.strip()

# 2) Trim strings ONLY in object (string-ish) columns to avoid dtype pollution
obj_cols = df.select_dtypes(include=["object"]).columns
df[obj_cols] = df[obj_cols].apply(lambda s: s.map(lambda v: v.strip() if isinstance(v, str) else v))

# 3) Normalize null tokens ONLY in object columns
NULL_TOKENS = {"", "-", "--", "---"}
df[obj_cols] = df[obj_cols].apply(
    lambda s: s.map(lambda v: None if (isinstance(v, str) and v.strip() in NULL_TOKENS) else v)
)


for col in DATETIME_COLS:
    if col in df.columns:
        parsed = pd.to_datetime(df[col], errors="coerce", utc=True)
        df[col] = parsed.dt.strftime("%Y-%m-%dT%H:%M:%SZ").where(parsed.notna(), None)
        df[col] = parsed.astype(object).where(parsed.notna(), None)
        

# 5) Add "MOC" prefix to mocId (if present)
if "mocId" in df.columns:
    def add_moc_prefix(x):
        if x is None:
            return None
        s = str(x).strip()
        if not s:
            return None
        return s if s.startswith("MOC") else f"MOC{s}"
    df["mocId"] = df["mocId"].map(add_moc_prefix)

# 6) Convert any remaining NaN/NaT to None (safe for Neo4j)
df = df.replace({np.nan: None})
df = df.where(pd.notna(df), None)



for col in INT_COLS:
    if col in df.columns:
        # 1) Convert anything numeric-looking to numbers; bad values like 'N' -> NaN
        s = pd.to_numeric(df[col], errors="coerce")

        # 2) OPTIONAL: if you only want whole numbers, drop decimals (12.5 -> NaN)
        # remove this line if decimals are valid for your column
        s = s.where(s.isna() | (s % 1 == 0), np.nan)

        # 3) Convert to nullable integer (supports missing values as <NA>)
        df[col] = s.astype("Int64")


rows = df.to_dict(orient="records")

# 10) Upload in batches
with driver.session(database=DATABASE) as session:
    for batch in chunked(rows, BATCH_SIZE):
        upload_batch(session, batch)

driver.close()
print("Upload complete ✅ (integers enforced in payload).")







```
